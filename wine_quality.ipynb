```python
# Fill missing values with the mean of the same class
for col in df3.columns[:-1]:  # Exclude the target_class column
    df3[col] = df3[col].fillna(df3.groupby('quality')[col].transform('mean'))
```

```python
import pandas as pd
import numpy as np
import imblearn
from sklearn.metrics import classification_report

from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

from imblearn.over_sampling import SMOTE

from sklearn.ensemble import RandomForestClassifier
from imblearn.combine import SMOTEENN
```

```python
#Dataset Imbalance
print(df2['quality'].value_counts())
fig = plt.figure(figsize = (9,2))
df2['quality'].value_counts(normalize = True).plot.barh()
```

```python
df=df2
```

```python
train, test = train_test_split(df, test_size = 0.20, random_state = 42, stratify = df['quality'])
X_train, y_train, X_test, y_test = train.drop(['quality'], axis = 1), train['quality'], \
                                   test.drop(['quality'], axis = 1), test['quality']
```

```python
y_train.shape, y_train.value_counts()
```

```python
# Initialize SMOTE with a specific random_state
smote_os = SMOTE(random_state=12,k_neighbors=3)
X_smote, y_smote = smote_os.fit_resample(X_train, y_train)
```

```python
# Initialize SMOTEENN with a specific random_state
smote_enn = SMOTEENN(random_state=12,smote=SMOTE(k_neighbors=3))
X_smoteenn, y_smoteenn = smote_enn.fit_resample(X_train, y_train)
```

```python
# Display value counts after applying SMOTE and SMOTEENN
print("Value counts after applying SMOTE:")
print(y_smote.value_counts())
```

```python
print("\nValue counts after applying SMOTEENN:")
print(y_smoteenn.value_counts())
```

```python
comparison_sets = [
    (X_train, y_train),
    (X_smote, y_smote),
    (X_smoteenn, y_smoteenn)
]
```

```python
# Initialize RandomForestClassifier
rf = RandomForestClassifier(random_state=12, max_depth=5)

# Train RandomForestClassifier and print classification reports
for i in range(len(comparison_sets)):
    print(f"\nResults for comparison set {i + 1}:")
    rf.fit(comparison_sets[i][0], comparison_sets[i][1])
    print(classification_report(y_test, rf.predict(X_test),zero_division=1))
```
